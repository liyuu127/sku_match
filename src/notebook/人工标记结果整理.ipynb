```python
# 读取408和92的人工核查结果，整理出对应附件2的商品信息
```

```python
# 加载人工标记文件408
import pandas as pd

df = pd.read_excel('../../data/附件1-人工标记商品.xlsx')
df
```

```python
df['tops_match'] = df[['top1_match', 'top2_match', 'top3_match']].apply(lambda x: 1 if x.any() else 0, axis=1)
# human_match_ID列不为空，NaN,human_match=1，否则为0
df['human_match'] = df['human_match_ID'].fillna('').astype(str).str.strip().apply(
    lambda x: 1 if str(x).strip() != '' else 0)
```

```python
# 提取df 本店商品ID、本店商品名称、相似商品1信息、top1_match,相似商品2信、top2_match,相似商品3信息、top3_match,tops_match,human_match
part_408_df = df[
    ['本店商品ID', '本店商品名称', '相似商品1信息', 'top1_match', '相似商品2信息', 'top2_match', '相似商品3信息',
     'top3_match', 'tops_match', 'human_match', 'human_match_ID']]
part_408_df
```

```python
# 重命名列本店商品ID为商品ID
part_408_df.rename(columns={'本店商品ID': '商品ID'}, inplace=True)
part_408_df.rename(columns={'本店商品名称': '商品名称'}, inplace=True)
part_408_df.rename(columns={'相似商品1信息': '相似商品1'}, inplace=True)
part_408_df.rename(columns={'相似商品2信息': '相似商品2'}, inplace=True)
part_408_df.rename(columns={'相似商品3信息': '相似商品3'}, inplace=True)
part_408_df.rename(columns={'human_match_ID': 'human_match_id'}, inplace=True)

part_408_df['human_match_id'] = part_408_df['human_match_id'].fillna('').astype(str).str.strip()
# tops_match ，填充空为0,转为 int
part_408_df['tops_match'] = part_408_df['tops_match'].fillna(0).astype(int)
part_408_df['top1_match'] = part_408_df['top1_match'].fillna(0).astype(int)
part_408_df['top2_match'] = part_408_df['top2_match'].fillna(0).astype(int)
part_408_df['top3_match'] = part_408_df['top3_match'].fillna(0).astype(int)
part_408_df['human_match'] = part_408_df['human_match'].fillna(0).astype(int)

part_408_df
```

```python
# 处理humatch_ID列
import re
import math


def pandas_str_to_series(s) -> pd.Series:
    """字符串转为Series"""
    # 判断是否已经是 Series
    s = str(s)
    s = s.strip()

    # s为None或""或nan
    if s.strip() in ["", "nan", "None", "NaN"]:
        return None

    inner = s[s.find("(") + 1: s.rfind(")")]

    pattern = re.compile(r"(\w+)=('[^']*'|[^,]*)")
    data = {k: v.strip("'") if v.strip() != "nan" else None for k, v in pattern.findall(inner)}

    # 3️⃣ 转为 DataFrame
    df = pd.DataFrame([data])
    return df.iloc[0]
```

```python
# 处理tops_match_id,human_match_id列
# match只代表是否匹配到，不代表召回率
for index, row in part_408_df.iterrows():
    human_match_id = ''
    top1_match_id = ''
    top2_match_id = ''
    top3_match_id = ''
    tops_match_id = ''
    human_match = 0
    # top3中是否存在
    if row['top1_match'] == 1:
        series = pandas_str_to_series(row['相似商品1'])
        if series is not None:
            human_match_id = series['商品ID']
            top1_match_id = series['商品ID']
            tops_match_id = series['商品ID']
            human_match = 1
    elif row['top2_match'] == 1:
        series = pandas_str_to_series(row['相似商品2'])
        if series is not None:
            human_match_id = series['商品ID']
            top2_match_id = series['商品ID']
            tops_match_id = series['商品ID']
            human_match = 1
    elif row['top3_match'] == 1:
        series = pandas_str_to_series(row['相似商品3'])
        if series is not None:
            human_match_id = series['商品ID']
            top3_match_id = series['商品ID']
            tops_match_id = series['商品ID']
            human_match = 1
    elif row['human_match_id'] != '':
        human_match_id = row['human_match_id']
        human_match = 1

    part_408_df.at[index, 'top1_match_id'] = top1_match_id
    part_408_df.at[index, 'top2_match_id'] = top2_match_id
    part_408_df.at[index, 'top3_match_id'] = top3_match_id
    part_408_df.at[index, 'tops_match_id'] = tops_match_id
    part_408_df.at[index, 'human_match'] = human_match
    part_408_df.at[index, 'human_match_id'] = human_match_id

part_408_df
```

```python
# 加载92条记录
```

```python
part_92_df = pd.read_excel('../../data/top3相似召回_glm9b_v3.xlsx')
part_92_df
```

```python
part_92_df = part_92_df[
    ['商品ID', '商品名称', '相似商品1', 'top1_匹配结果', '相似商品2', 'top2_匹配结果', '相似商品3',
     'top3_匹配结果', 'tops_匹配结果', '商品ID_y']]
part_92_df.rename(columns={'top1_匹配结果': 'top1_match'}, inplace=True)
part_92_df.rename(columns={'top2_匹配结果': 'top2_match'}, inplace=True)
part_92_df.rename(columns={'top3_匹配结果': 'top3_match'}, inplace=True)
part_92_df.rename(columns={'tops_匹配结果': 'tops_match'}, inplace=True)
part_92_df.rename(columns={'商品ID_y': 'human_match_id'}, inplace=True)
part_92_df['human_match'] = 1

# part_92_df 相似商品1列 填充空为'' 并转为str
part_92_df['相似商品1'] = part_92_df['相似商品1'].fillna('').astype(str).str.strip()
part_92_df['相似商品2'] = part_92_df['相似商品2'].fillna('').astype(str).str.strip()
part_92_df['相似商品3'] = part_92_df['相似商品3'].fillna('').astype(str).str.strip()

part_92_df['human_match_id'] = part_92_df['human_match_id'].fillna('').astype(str).str.strip()
part_92_df['tops_match'] = part_92_df['tops_match'].fillna(0).astype(int)
part_92_df['top1_match'] = part_92_df['top1_match'].fillna(0).astype(int)
part_92_df['top2_match'] = part_92_df['top2_match'].fillna(0).astype(int)
part_92_df['top3_match'] = part_92_df['top3_match'].fillna(0).astype(int)
part_92_df['human_match'] = part_92_df['human_match'].fillna(0).astype(int)
part_92_df
```

```python
for index, row in part_92_df.iterrows():
    top1_match_id = ''
    top2_match_id = ''
    top3_match_id = ''
    tops_match_id = ''
    # top3中是否存在
    if row['top1_match'] == 1:
        series = pandas_str_to_series(row['相似商品1'])
        if series is not None:
            top1_match_id = series.商品ID
            tops_match_id = series['商品ID']
    elif row['top2_match'] == 1:
        series = pandas_str_to_series(row['相似商品2'])
        if series is not None:
            top2_match_id = series['商品ID']
            tops_match_id = series['商品ID']
    elif row['top3_match'] == 1:
        series = pandas_str_to_series(row['相似商品3'])
        if series is not None:
            top3_match_id = series['商品ID']
            tops_match_id = series['商品ID']

    part_92_df.at[index, 'top1_match_id'] = top1_match_id
    part_92_df.at[index, 'top2_match_id'] = top2_match_id
    part_92_df.at[index, 'top3_match_id'] = top3_match_id
    part_92_df.at[index, 'tops_match_id'] = tops_match_id
part_92_df
```
